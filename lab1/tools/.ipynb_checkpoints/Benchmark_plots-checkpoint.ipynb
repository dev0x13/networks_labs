```python
%matplotlib notebook

import matplotlib.pyplot as plt
```

```python
import csv
import numpy as np
```

```python
loss_prob_GBN = []
loss_prob_SR = []
window_size_GBN = []
window_size_SR = []

with open("loss_prob_GBN.csv", "r") as log:
    for r in csv.reader(log, delimiter=","):
        loss_prob_GBN.append((float(r[0]), int(r[1]), int(r[2]), float(r[3]), float(r[4])))
    #loss_prob_GBN.append((1, int(r[1]), 0, 0, np.inf))
    
with open("loss_prob_SR.csv", "r") as log:
    for r in csv.reader(log, delimiter=","):
        loss_prob_SR.append((float(r[0]), int(r[1]), int(r[2]), float(r[3]), float(r[4])))
    #loss_prob_SR.append((1, int(r[1]), 0, 0, np.inf))

with open("window_size_GBN.csv", "r") as log:
    for r in csv.reader(log, delimiter=","):
        window_size_GBN.append((int(r[0]), int(r[1]), int(r[2]), float(r[3]), float(r[4])))

with open("window_size_SR.csv", "r") as log:
    for r in csv.reader(log, delimiter=","):
        window_size_SR.append((int(r[0]), int(r[1]), int(r[2]), float(r[3]), float(r[4])))
```

```python
loss_prob_GBN = np.array(loss_prob_GBN)
loss_prob_SR = np.array(loss_prob_SR)
window_size_GBN = np.array(window_size_GBN)
window_size_SR = np.array(window_size_SR)
```

```python
plt.figure()

plt.plot(loss_prob_GBN[:, 0], loss_prob_GBN[:, 3], "ro-")
plt.plot(loss_prob_SR[:, 0], loss_prob_SR[:, 3], "gd-")

plt.ylim(0, 1)
plt.xlim(0, 1)
plt.legend(["Go-Back-N", "Selective Repeat"])
plt.xlabel("Вероятность потери пакета")
plt.ylabel("Коэффициент эффективности")
```

```python
plt.figure()

plt.plot(loss_prob_GBN[:, 0], loss_prob_GBN[:, 4], "ro-")
plt.plot(loss_prob_SR[:, 0], loss_prob_SR[:, 4], "gd-")

plt.xlim(0, 1)
plt.legend(["Go-Back-N", "Selective Repeat"])
plt.xlabel("Вероятность потери пакета")
plt.ylabel("Время передачи, мс")
```

```python
plt.figure()

plt.plot(window_size_GBN[:, 0], window_size_GBN[:, 3], "ro-")
plt.plot(window_size_SR[:, 0], window_size_SR[:, 3], "gd-")

plt.ylim(0, 0.8)
plt.xlim(1, 10)
plt.legend(["Go-Back-N", "Selective Repeat"])
plt.xlabel("Размер окна")
plt.ylabel("Коэффициент эффективности")
```

```python
plt.figure()

plt.plot(window_size_GBN[:, 0], window_size_GBN[:, 4], "ro-")
plt.plot(window_size_SR[:, 0], window_size_SR[:, 4], "gd-")

plt.xlim(1, 10)
plt.legend(["Go-Back-N", "Selective Repeat"])
plt.xlabel("Размер окна")
plt.ylabel("Время передачи, мс")
```
